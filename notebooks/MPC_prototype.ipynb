```python
# =========================================================
# üß† AN√ÅLISE EXPLORAT√ìRIA DE DADOS (EDA)
# Projeto: Otimiza√ß√£o de Consumo Energ√©tico em Data Centers
# Dataset: cold_source_control_dataset.csv
# =========================================================

# 1Ô∏è‚É£ Importa√ß√£o das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import os

# 2Ô∏è‚É£ Defini√ß√£o do caminho do arquivo (ajuste conforme necess√°rio)
file_path = "C:\\Users\\rafae\\Desktop\\data-center-energy-optimizer\\data\\raw\\cold_source_control_dataset.csv"

# 3Ô∏è‚É£ Verificar se o arquivo existe no diret√≥rio
if os.path.exists(file_path):
    print("‚úÖ Arquivo encontrado com sucesso!")
else:
    print("‚ùå Caminho incorreto! Verifique o diret√≥rio e o nome do arquivo.")

# 4Ô∏è‚É£ Leitura segura do CSV
df = pd.read_csv(file_path)

# 5Ô∏è‚É£ Exibir informa√ß√µes gerais do dataset
print("\nüìã Informa√ß√µes gerais:")
print(df.info())

# 6Ô∏è‚É£ Exibir as 5 primeiras linhas
print("\nüîç Primeiras linhas do dataset:")
display(df.head())

# 7Ô∏è‚É£ Verificar valores ausentes
print("\nüß© Valores ausentes por coluna:")
print(df.isnull().sum())

# 8Ô∏è‚É£ Estat√≠sticas descritivas b√°sicas
print("\nüìä Estat√≠sticas descritivas:")
display(df.describe())

# 9Ô∏è‚É£ Plotar vis√£o geral das principais vari√°veis num√©ricas
numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns
df[numeric_cols].hist(bins=20, figsize=(12,10))
plt.suptitle("Distribui√ß√£o das Vari√°veis Num√©ricas", fontsize=14)
plt.show()
```

```python
# =========================================================
# ‚öôÔ∏è PARTE 2 ‚Äî AN√ÅLISE TEMPORAL E C√ÅLCULO DO PUE
# =========================================================

import numpy as np

# 1Ô∏è‚É£ Converter a coluna Timestamp para formato de data/hora (caso ainda n√£o esteja)
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# 2Ô∏è‚É£ Ordenar o dataset cronologicamente
df = df.sort_values(by='Timestamp')

# 3Ô∏è‚É£ Visualizar s√©rie temporal das temperaturas e carga do servidor
plt.figure(figsize=(14,6))
plt.plot(df['Timestamp'], df['Inlet_Temperature(¬∞C)'], label='Inlet Temp (¬∞C)', linewidth=1)
plt.plot(df['Timestamp'], df['Outlet_Temperature(¬∞C)'], label='Outlet Temp (¬∞C)', linewidth=1)
plt.plot(df['Timestamp'], df['Ambient_Temperature(¬∞C)'], label='Ambient Temp (¬∞C)', linewidth=1, linestyle='--')
plt.title("üìâ S√©ries Temporais de Temperatura")
plt.xlabel("Tempo")
plt.ylabel("Temperatura (¬∞C)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14,5))
plt.plot(df['Timestamp'], df['Server_Workload(%)'], color='darkorange', label='Server Workload (%)')
plt.title("‚öôÔ∏è Carga de Trabalho dos Servidores ao Longo do Tempo")
plt.xlabel("Tempo")
plt.ylabel("Carga (%)")
plt.legend()
plt.grid(True)
plt.show()

# 4Ô∏è‚É£ Estimar o consumo de TI e calcular o PUE
# Supondo que o consumo de TI (IT Power) √© proporcional √† carga de trabalho dos servidores
df['IT_Power(kW)'] = df['Server_Workload(%)'] / 100 * df['Cooling_Unit_Power_Consumption(kW)'].max()

# C√°lculo do PUE (Power Usage Effectiveness)
# PUE = (Total Facility Power) / (IT Power)
df['PUE'] = (df['Cooling_Unit_Power_Consumption(kW)']) / (df['IT_Power(kW)'])
df['PUE'] = df['PUE'].replace([np.inf, -np.inf], np.nan)

# 5Ô∏è‚É£ Estat√≠sticas do PUE
print("\n‚ö° Estat√≠sticas do PUE (Power Usage Effectiveness):")
display(df['PUE'].describe())

# 6Ô∏è‚É£ Visualiza√ß√£o do PUE
plt.figure(figsize=(12,5))
plt.plot(df['Timestamp'], df['PUE'], color='purple', label='PUE')
plt.title("üìä Evolu√ß√£o Temporal do PUE (Efici√™ncia Energ√©tica)")
plt.xlabel("Tempo")
plt.ylabel("PUE")
plt.legend()
plt.grid(True)
plt.show()

# 7Ô∏è‚É£ Distribui√ß√£o do PUE
plt.figure(figsize=(8,5))
plt.hist(df['PUE'].dropna(), bins=20, color='mediumpurple', edgecolor='black')
plt.title("Distribui√ß√£o do PUE")
plt.xlabel("PUE")
plt.ylabel("Frequ√™ncia")
plt.show()

# 8Ô∏è‚É£ Valor m√©dio e interpreta√ß√£o r√°pida
pue_medio = df['PUE'].mean()
print(f"\nüîé PUE m√©dio estimado: {pue_medio:.2f}")

if pue_medio <= 1.5:
    print("‚úÖ Excelente efici√™ncia ‚Äî desempenho similar a data centers de alta performance.")
elif 1.5 < pue_medio <= 2.0:
    print("‚öôÔ∏è Efici√™ncia moderada ‚Äî h√° margem para otimiza√ß√£o do sistema de refrigera√ß√£o.")
else:
    print("‚ö†Ô∏è Baixa efici√™ncia ‚Äî recomenda-se otimiza√ß√£o energ√©tica urgente (ex: controle inteligente de resfriamento).")
```

```python
# =========================================================
# üîç PARTE 3 ‚Äî CORRELA√á√ïES, INEFICI√äNCIAS E BASE DE CONTROLE
# =========================================================

import seaborn as sns
import numpy as np

# 1Ô∏è‚É£ Selecionar apenas as vari√°veis num√©ricas relevantes
cols_interesse = [
    'Server_Workload(%)', 
    'Inlet_Temperature(¬∞C)',
    'Outlet_Temperature(¬∞C)',
    'Ambient_Temperature(¬∞C)',
    'Cooling_Unit_Power_Consumption(kW)',
    'Chiller_Usage(%)',
    'AHU_Usage(%)',
    'Total_Energy_Cost($)',
    'Temperature_Deviation(¬∞C)',
    'PUE'
]

df_corr = df[cols_interesse].corr()

# 2Ô∏è‚É£ Heatmap de correla√ß√µes
plt.figure(figsize=(10,8))
sns.heatmap(df_corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Mapa de Correla√ß√£o das Vari√°veis")
plt.show()

# 3Ô∏è‚É£ Identificar per√≠odos de baixa efici√™ncia (PUE alto)
limite_ineficiencia = df['PUE'].quantile(0.9)
ineficientes = df[df['PUE'] > limite_ineficiencia]

print(f"\n‚ö†Ô∏è Total de registros com PUE alto (> {limite_ineficiencia:.2f}): {len(ineficientes)}")

# Exibir amostra dos per√≠odos cr√≠ticos
display(ineficientes[['Timestamp', 'PUE', 'Server_Workload(%)', 
                      'Cooling_Unit_Power_Consumption(kW)', 
                      'Ambient_Temperature(¬∞C)', 
                      'Temperature_Deviation(¬∞C)']].head(10))

# 4Ô∏è‚É£ Gr√°fico de compara√ß√£o entre per√≠odos eficientes e ineficientes
plt.figure(figsize=(12,6))
plt.scatter(df['Server_Workload(%)'], df['PUE'], c='lightgray', label='Normal')
plt.scatter(ineficientes['Server_Workload(%)'], ineficientes['PUE'], c='red', label='Ineficiente')
plt.title("Rela√ß√£o entre Carga de Trabalho e PUE")
plt.xlabel("Server Workload (%)")
plt.ylabel("PUE")
plt.legend()
plt.grid(True)
plt.show()

# 5Ô∏è‚É£ Base inicial para controle inteligente de refrigera√ß√£o
# Estrat√©gia: ajustar o resfriamento conforme a carga e o desvio t√©rmico
def estrategia_resfriamento(workload, temp_dev, pue):
    if pue > 2.0 and temp_dev > 2:
        return "Aumentar efici√™ncia do chiller (otimizar fluxo de ar)"
    elif pue > 1.5 and workload < 50:
        return "Reduzir velocidade dos ventiladores (modo econ√¥mico)"
    elif temp_dev < -1:
        return "Reduzir resfriamento ‚Äî overcooling detectado"
    else:
        return "Opera√ß√£o normal"

df['Cooling_Strategy_Recommendation'] = df.apply(
    lambda row: estrategia_resfriamento(
        row['Server_Workload(%)'],
        row['Temperature_Deviation(¬∞C)'],
        row['PUE']
    ), axis=1
)

print("\nüí° Recomenda√ß√µes autom√°ticas de controle (amostra):")
display(df[['Timestamp', 'Server_Workload(%)', 'PUE', 'Temperature_Deviation(¬∞C)', 'Cooling_Strategy_Recommendation']].head(10))
```

```python
# =========================================================
# PARTE 4 ‚Äî PROT√ìTIPO: identificar din√¢mica linear + MPC
# =========================================================

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import cvxpy as cp
import matplotlib.pyplot as plt

# ---------- 1. Preparar dados (assume df j√° carregado e Timestamp parseado) ----------
df = df.copy()
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df = df.sort_values('Timestamp').reset_index(drop=True)

# Para simplificar: renomear colunas curtas
df.rename(columns={
    "Server_Workload(%)":"workload",
    "Inlet_Temperature(¬∞C)":"inlet_temp",
    "Ambient_Temperature(¬∞C)":"ambient_temp",
    "Cooling_Unit_Power_Consumption(kW)":"cooling_power",
    "Temperature_Deviation(¬∞C)":"temp_dev"
}, inplace=True)

# Se existir coluna de IT power / total power usar; aqui usamos cooling_power as control proxy
# Cria delta_t = intervalo em horas (assume uniform):
# se timestamps irregulares, recompute dt_series = df['Timestamp'].diff().dt.total_seconds()/3600
dt_hours = (df['Timestamp'].diff().dt.total_seconds().fillna(method='bfill') / 3600.0).median()

# ---------- 2. Construir dados para identifica√ß√£o (modelo simples)
# modelo assumido: inlet_{t+1} = a*inlet_t + b*cooling_action_t + c*ambient_t + d*workload_t + e
# aqui usamos cooling_power (registrado) como proxy da a√ß√£o; iremos depois otimizar sobre uma vari√°vel similar.

# Build X, y for one-step-ahead regression
X_list, y_list = [], []
for i in range(len(df)-1):
    Xi = [ df.loc[i, 'inlet_temp'],
           df.loc[i, 'cooling_power'],     # proxy action
           df.loc[i, 'ambient_temp'],
           df.loc[i, 'workload'] ]
    yi = df.loc[i+1, 'inlet_temp']
    if not np.any(pd.isnull(Xi)) and not pd.isnull(yi):
        X_list.append(Xi)
        y_list.append(yi)

X = np.array(X_list)
y = np.array(y_list)

# Fit linear regression
lr = LinearRegression()
lr.fit(X, y)
a, b, c, d = lr.coef_
e = lr.intercept_
print("Modelo identificado: inlet_next = a*inlet + b*cooling + c*ambient + d*workload + e")
print(f"a={a:.4f}, b={b:.4f}, c={c:.4f}, d={d:.4f}, e={e:.4f}")

# ---------- 3. MPC setup ----------
# Decision variable: cooling_action (kW) at each step in horizon H
H = 6  # horizonte (ex: 6 passos de 15 min => 1.5h); ajuste conforme freq do seu dataset
# constraints, tuning
cooling_min = 0.3   # kW, m√≠nimo seguro (ajuste com base nos dados)
cooling_max = 1.2   # kW, m√°ximo do equipamento (ajuste)
Tmax = 25.0         # temperatura m√°xima aceit√°vel no inlet (¬∞C) ‚Äî altere conforme ASHRAE/objetivo
alpha = 1.0         # peso energ√©tico (cost per kW of cooling)
beta = 50.0         # penaliza√ß√£o por violar temperatura (high penalty)

# initialize arrays to store results
n = len(df)
mpc_actions = np.full(n, np.nan)
mpc_pred_inlet = np.full(n, np.nan)

# For simulation we will use measured ambient and workload as exogenous (assumed known/predicted)
ambient = df['ambient_temp'].values
workload = df['workload'].values
meas_inlet = df['inlet_temp'].values
baseline_cooling = df['cooling_power'].values  # baseline to compare energy

# We'll run MPC in receding horizon, using the linear model to predict next states
for t in range(n - H - 1):
    # current state
    inlet_t = float(meas_inlet[t])
    # build predicted exogenous over horizon (use measured future ambient/workload as "forecast")
    amb_forecast = ambient[t+1:t+1+H]
    wl_forecast = workload[t+1:t+1+H]

    # cvxpy variables
    u = cp.Variable(H)           # cooling actions to decide (kW)
    T = cp.Variable(H)           # predicted inlet temperatures (¬∞C)

    constraints = []
    # dynamic constraints
    for k in range(H):
        if k == 0:
            # T[0] = a*inlet_t + b*u[0] + c*amb_forecast[0] + d*wl_forecast[0] + e
            constraints.append(T[0] == a*inlet_t + b*u[0] + c*amb_forecast[0] + d*wl_forecast[0] + e)
        else:
            # T[k] = a*T[k-1] + b*u[k] + c*amb_forecast[k] + d*wl_forecast[k] + e
            constraints.append(T[k] == a*T[k-1] + b*u[k] + c*amb_forecast[k] + d*wl_forecast[k] + e)
        # action limits
        constraints.append(u[k] >= cooling_min)
        constraints.append(u[k] <= cooling_max)
        # optionally bound temperatures to physically plausible range
        constraints.append(T[k] >= 10.0)
        constraints.append(T[k] <= 50.0)

    # objective: minimize sum(alpha*u + beta*relu(T-Tmax)^2)
    # implement soft constraint via quadratic penalty on exceedance
    exceed = cp.pos(T - Tmax)  # positive part
    objective = cp.sum(alpha * u + beta * cp.sum_squares(exceed))
    prob = cp.Problem(cp.Minimize(objective), constraints)

    # solve
    try:
        prob.solve(solver=cp.OSQP, warm_start=True, verbose=False)
    except Exception as exc:
        # fallback: try default solver
        prob.solve()

    # get action to apply (first element)
    if u.value is not None:
        action = float(u.value[0])
        mpc_actions[t] = action
        # store predicted T[0] for visualization
        mpc_pred_inlet[t] = float(T.value[0])
    else:
        # solver failed ‚Äî fall back to baseline
        mpc_actions[t] = baseline_cooling[t]
        mpc_pred_inlet[t] = inlet_t

# ---------- 4. Avalia√ß√£o ----------
# Compare total cooling energy (kW * dt_hours) over simulation period
valid_idx = ~np.isnan(mpc_actions)
energy_mpc = np.nansum(mpc_actions[valid_idx]) * (dt_hours)   # kWh approx
energy_baseline = np.nansum(baseline_cooling[valid_idx]) * (dt_hours)

print(f"Consumo de refrigera√ß√£o (baseline): {energy_baseline:.2f} kWh")
print(f"Consumo de refrigera√ß√£o (MPC):      {energy_mpc:.2f} kWh")
if energy_baseline>0:
    print(f"Redu√ß√£o relativa: {(energy_baseline-energy_mpc)/energy_baseline*100:.2f} %")

# Violations: times when measured inlet temp > Tmax (baseline) and predicted by MPC sim
viol_baseline = np.sum(meas_inlet[valid_idx] > Tmax)
# For MPC we can simulate the inlet using the surrogate model + applied actions to check T evolution
sim_inlet = meas_inlet.copy()
for t in np.where(valid_idx)[0]:
    # apply model one-step with mpc_actions
    if t+1 < len(sim_inlet):
        sim_inlet[t+1] = a*sim_inlet[t] + b*mpc_actions[t] + c*ambient[t+1] + d*workload[t+1] + e

viol_mpc = np.sum(sim_inlet[valid_idx] > Tmax)

print(f"Violations baseline (measured inlet > {Tmax}¬∞C): {viol_baseline}")
print(f"Violations simulated MPC (pred inlet > {Tmax}¬∞C): {viol_mpc}")

# ---------- 5. Plots ----------
plt.figure(figsize=(12,4))
plt.plot(df['Timestamp'], baseline_cooling, label='Baseline cooling (kW)', alpha=0.7)
plt.plot(df['Timestamp'], mpc_actions, label='MPC action (kW)', alpha=0.7)
plt.title('Baseline vs MPC cooling actions')
plt.legend()
plt.show()

plt.figure(figsize=(12,4))
plt.plot(df['Timestamp'], meas_inlet, label='Measured inlet (¬∞C)', alpha=0.7)
plt.plot(df['Timestamp'], mpc_pred_inlet, label='MPC predicted inlet next (¬∞C)', alpha=0.7)
plt.axhline(Tmax, color='r', linestyle='--', label=f'Tmax = {Tmax}¬∞C')
plt.title('Inlet temperature: measured vs MPC prediction')
plt.legend()
plt.show()
```
